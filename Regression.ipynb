# Ejemplo de regresión

```python
# Needed libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error
```

## Visualización del dataset

```python
# Dataset creation for regresion
X, y = make_regression(n_samples=500, n_features=1, noise=30)

# Plot outputs
plt.scatter(X, y, color="c")
plt.title('Regression data')
plt.xlabel('x value')
plt.ylabel('y value')

plt.show()
```

## Split del conjunto de prueba y entrenamiento

```python
# Select the split threshold
split = int(0.2*X.shape[0])

# Split the data into training/testing sets
X_train = X[:-split]
X_test = X[-split:]

# Split the targets into training/testing sets
y_train = y[:-split]
y_test = y[-split:]

# Linspace for regression line evaluation
x_axis = np.linspace(X_train.min(), X_train.max(), num=X_train.shape[0])
x_axis = x_axis.reshape(x_axis.shape[0],1)
```

## Entrenamiento y evaluación del modelo

```python
# Create linear regression model
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)

# Plot outputs
plt.scatter(X_train, y_train, color="r", label='Train data')
plt.scatter(X_test, y_test, color="b", label='Test data')
plt.plot(x_axis, regr.predict(x_axis), color="c", linewidth=3, label='Regression line')

plt.title('Regression data')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend()

plt.show()
```

```python
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```

## Train and evaluation of sorted data

```python
# For sorted data
Data = list(zip(X, y))
Data.sort()
a, b = zip(*Data)
X, y, = np.array(a), np.array(b)


# Select the split threshold
split = int(0.2*X.shape[0])

# Split the data into training/testing sets
X_train = X[:-split]
X_test = X[-split:]

# Split the targets into training/testing sets
y_train = y[:-split]
y_test = y[-split:]



# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(X_test)



# Plot outputs
plt.scatter(X_train, y_train, color="r", label='Train data')
plt.scatter(X_test, y_test, color="b", label='Test data')
plt.plot(x_axis, regr.predict(x_axis), color="c", linewidth=3, label='Regression line')

plt.title('Regression data')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend()

plt.show()
```

```python
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```
